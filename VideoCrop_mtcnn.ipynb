```python
import cv2
import glob
import os
import dlib
import numpy as np
from facenet_pytorch import MTCNN
from IQA_pytorch import SSIM
```

```python
img_size=256
sample_frame_num = 30

def video2frame(invideofilename, save_path):
    vidcap = cv2.VideoCapture(invideofilename)
    videoname = os.path.basename(invideofilename).split('.')[0]
    count = 0
    save_count=0

    while True:
        success,image = vidcap.read()
        if not success:
            break
        mtcnn=MTCNN(image_size=img_size,margin=20,keep_all=True)
        faces= mtcnn(image)
#         print ('Read a new frame: ', success)
        
        if faces==None:
            print('noface')
            count+=1
            continue
                    
        elif len(faces) >= 1 and save_count > 0:
            mse_score = []
            for f in faces:
                score = -np.sum((prev_image.astype("float") - np.asarray(f).astype("float")) ** 2)/img_size**2
                mse_score.append(score)
            if np.abs(np.argmax(mse_score)) > 0.5:
                count+=1
                continue

            face_img = faces[np.argmax(mse_score)]

        elif save_count==0:
            face_img=faces[0]

        faces_np = face_img.permute(1, 2, 0).numpy()
        face = cv2.normalize(faces_np,None,0,255,cv2.NORM_MINMAX)
        fname = "{}_{}.jpg".format(videoname,"{0:05d}".format(count))
        cv2.imwrite(save_path + fname, face) # save frame as JPEG file
        save_count+=1

        prev_image = np.asarray(face_img)
        
        count += 1
        
        if save_count==90:
            return count
            break

    print("{} images are extracted in {}.". format(count, save_path))
```
